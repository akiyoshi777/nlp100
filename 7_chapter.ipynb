# ７章

Google Newsデータセット(1000億単語)で学習済み単語ベクトルを使用

```python
from gensim.models import KeyedVectors
model = KeyedVectors.load_word2vec_format('data/GoogleNews-vectors-negative300.bin.gz', binary=True)
```

### 60. 単語ベクトルの読み込みと表示

```python
# United_Statesの単語ベクトル
# 300次元の確認
print(model['Unites_States'].shape)
model['United_States']
```

### 61. 単語の類似度

```python
# United_StatesとU.Sのコサイン類似度
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
# cos類似度関数を作成
def cossim(a,b):
    return np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b))

print('cosine_similarity関数使用')
print(cosine_similarity(model['United_States'].reshape(1, -1), model['U.S.'].reshape(1 ,-1))) # 1行適当な大きさの列のベクトルに変更
print('自作のcos類似度関数使用')
print(cossim(model['United_States'], model['U.S.']))
```

```python
cossim(model['United_States'], model['U.S.'])
```

### 62. 類似度の高い単語10件

```python
# United_Statesとコサイン類似度が高い10語
result = model.most_similar(positive=['United_States'], topn=10)
```

```python
for item in result:
    print(f'{item[0]}:{item[1]:.4f}') # 小数点4桁まで表示
```

### 63. 加法構成性によるアナロジー

```python
# "Spain"の単語ベクトルから"Madrid"のベクトルを引き，"Athens"のベクトルを足したベクトルを計算し，そのベクトルと類似度の高い10語とその類似度を出力せよ．
target = model['Spain']-model['Madrid']+model['Athens']
result = model.most_similar(positive=target, topn=10)
for i in result:
  print(f'{i[0]}:{i[1]:.4f}')
```

```python
# 答えが上と違う．解答はこっち．どうしよう
result1 = model.most_similar(positive=['Spain', 'Athens'], negative=['Madrid'], topn=10)
for i in result1:
  print(f'{i[0]}:{i[1]:.4f}')
```
